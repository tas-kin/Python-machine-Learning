# Customization

```python
#Axis label
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
plt.plot(year,pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
#title
plt.title('World population projection')
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
plt.plot(year,pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
#ticks
plt.yticks([0,2,4,6,8,10])
plt.title('World population projection')
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
plt.plot(year,pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
#ticks 2
plt.title('World population projection')
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
plt.scatter(year,pop)
plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks([0,2,4,6,8,10],['0','2B','4B','6B','8B','10B'])
plt.show()
```

```python
#sizes
np_pop=np.array(pop)
np_pop=np_pop*100
plt.scatter(year,pop,s=np_pop) #argument s for size
plt.title('World population projection')
plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks([0,2,4,6,8,10],['0','2B','4B','6B','8B','10B'])
plt.show()
```

```python
#color
np_pop=np.array(pop)
np_pop=np_pop*100
plt.scatter(year,pop,s=np_pop,c='green') #argument c for color
plt.title('World population projection')
plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks([0,2,4,6,8,10],['0','2B','4B','6B','8B','10B'])
plt.show()
```

```python
#transparecy of the bubbles
plt.scatter(year,pop,s=np_pop,c='green',alpha=0.2) #alpha for transparency 0<=alpha<=1
plt.title('World population projection')
plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks([0,2,4,6,8,10],['0','2B','4B','6B','8B','10B'])
plt.show()
```

```python
#Additional customaization
#text 
plt.text(1950,2.519,'Bangladesh')
#grid
plt.grid(True)
plt.scatter(year,pop,s=np_pop,c='green',alpha=0.8) #alpha for transparency 0<=alpha<=1
plt.title('World population projection')
plt.xlabel('Year')
plt.ylabel('Population')
plt.yticks([0,2,4,6,8,10],['0','2B','4B','6B','8B','10B'])
plt.show()
```

# Histogram

```python
values = [10,8,6,4,2,1]
plt.hist(values,bins=5)
plt.show()
plt.clf() #cler one histogram and show another one
plt.hist(values,bins=5,rwidth=3)
plt.show()
```

```python
#Barchart
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]

plt.bar(y_pos, performance, align='center', alpha=0.8)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')
plt.show()
```

```python
#horizontal barchart
values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.barh(values,values,align='center')
plt.show()
```

```python
#plotting sideways back to back graph using barchart
import numpy as np
import matplotlib.pyplot as plt

women_pop = np.array([5., 30., 45., 22.])
men_pop     = np.array( [5., 25., 50., 20.])
X = np.arange(4)
plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'b')
plt.show()
```
